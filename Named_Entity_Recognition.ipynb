```python
sentence = """The Eiffel Tower was built from 1887 to 1889
by french engineer gustave eiffel, whose company
specialized in building metal frameworks and structures. """

"""
Person Eg: atik
place or location eg: bangladesh
Date eg: february, 2/6/2026
Time eg: 6:42
Money eg: 1 milliion dollar
organization eg: INeuron Private Limited
Percent eg: 20% , twenty percent


"""
```

```python
import nltk
nltk.word_tokenize(sentence)
```

```python
words = nltk.word_tokenize(sentence)
```

```python
nltk.pos_tag(words)
```

```python
tag_elements = nltk.pos_tag(words)
```
