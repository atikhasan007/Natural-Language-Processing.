### Wordnet lemmatizer
Lemmatization technique is like stemming. The output we will get after lemmatization is called lemma , which is a root word rather than root stem , the output of stemming. After lemmatization,
we will be getting a valid word that means the same thing


NLTK provides WordNetLemmatizer class which is a thin wrappper around the wordnet corpus . This class uses morphy() function to the wordnet corpusReader class to find  a lemma . let us understand it with an example -

## Q&A , chatbots, text summarization

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
'''
POS - Nonu-n
verb -v
adjective -a
adverb -r
'''
lemmatizer.lemmatize('going')
```

```python
lemmatizer.lemmatize('going', pos='n')
```

```python
lemmatizer.lemmatize('going', pos='v')
```

```python
lemmatizer.lemmatize('going', pos='a')
```

```python
lemmatizer.lemmatize('going', pos='r')
```

```python
words = ['eating', 'eats', 'eat', 'ate', 'adjustable', 'rafting', 'ability', 'meeting']
```

```python
for word in words:
  print(word+'=====>'+ lemmatizer.lemmatize(word, pos='n'))
```

```python
for word in words:
  print(word+'=====>'+ lemmatizer.lemmatize(word, pos='v'))
```

```python
lemmatizer.lemmatize('goes', pos='v')
```

```python
lemmatizer.lemmatize('fairly', pos='v'), lemmatizer.lemmatize('sportingly')
```

| বিষয়          | **Stemming**          | **Lemmatization**      |
| ------------- | --------------------- | ---------------------- |
| পদ্ধতি        | Rule-based কাটাকাটি   | Dictionary + POS based |
| Speed         | খুব দ্রুত             | তুলনামূলক ধীর          |
| Output        | অনেক সময় real word না | সবসময় real word        |
| Context বোঝে? | ❌ না                  | ✅ হ্যাঁ                |
| Accuracy      | কম                    | বেশি                   |
| Use case      | TF-IDF, BoW           | DL, BERT, NLP apps     |
